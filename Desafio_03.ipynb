```python
"""
b.	leia do teclado o número total de eleitores de um município, o número de 
votos brancos, nulos e válidos. Calcular e escrever o percentual que cada um 
representa em relação ao total de eleitores. 
(Valores inválidos: números negativos).
"""
from google.colab import output
x = 0;
while x != 1 :
  brancos = float(input("Entre com o valor dos votos BRANCOS: "))
  if brancos < 0:
    output.clear()
    print("*** Erro! Digite um numero maior ou igual a ZERO! ***")
  else:
    x = 1;
x = 0;
while x != 1 :
  nulos = float(input("Entre com o valor dos votos NULOS: "))
  if nulos < 0:
    output.clear()
    print("*** Erro! Digite um numero maior ou igual a ZERO! ***")
  else:
    x = 1;
x = 0;
while x != 1 :
  validos = float(input("Entre com o valor dos votos VALIDOS: "))
  if validos < 0:
    output.clear()
    print("*** Erro! Digite um numero maior ou igual a ZERO! ***")
  else:
    x = 1;
x = 0;
totalVotos = brancos + nulos + validos
percBrancos = (brancos / totalVotos) * 100
percNulos = (nulos / totalVotos) * 100
percValidos = (validos / totalVotos) * 100
print("============================================================")
print("================== Resultado da Eleição ====================")
print("============================================================")
print("Percentual de Votos Brancos: %.2f"% percBrancos,"%")
print("Percentual de Votos Nulos: %.2f"% percNulos,"%")
print("Percentual de Votos Validos: %.2f"% percValidos,"%")
print("Total dos Votos apurados: %d"%totalVotos,"Votos")
print("============================================================")
```

```python
#Nível Avançado
"""
a.	leia 3 valores (A, B e C) representando as medidas de um triângulo. Mostrar se tais medidas formam um 
triângulo ou não. Lembre-se que o valor de cada lado deve ser menor que a soma dos outros 2 lados e deve 
ser maior que o valor absoluto da diferença entre os outros 2 lados.
| B - C | <  A  < B + C
| A - C | <  B  < A + C
| A - B | <  C  < A + B
"""
A = int(input("Entre com o valor da medida A: "))
B = int(input("Entre com o valor da medida B: "))
C = int(input("Entre com o valor da medida C: "))
print("===================================================")
if A > (B-C) and A < (B+C):
  print("O numero A digitado forma o tringulo!")
else:
  print("O numero A digitado NÃO forma o tringulo!")
if B > (A-C) and B < (A+C):
  print("O numero B digitado forma o tringulo!")
else:
  print("O numero B digitado NÃO forma o tringulo!")
if C > (A-B) and C < (A+B):
  print("O numero C digitado forma o tringulo!")
else:
  print("O numero C digitado NÃO forma o tringulo!")
```

```python
"""
6.	Crie um código utilizando LAÇO WHILE:
a.	Para a7=0 enquanto a7 menor que 10, incrementando 1, insira um outro while 
aninhado internamente contendo para a8=0 até a8 menor que 10, incrementando 1, 
mostre na tela "(a7,a8)".
"""
a7 = 0;
print("==========================================================")
print("WHILE")
print("==========================================================")
while a7 < 10 :
  a8 = 0;
  while a8 < 10 :
    print("(",a7,",",a8,")")
    a8+=1
  a7+=1
print("==========================================================")
print("FOR")
print("==========================================================")
for a7 in range(0,10,1):
  for a8 in range(0,10,1):
    print("(",a7,",",a8,")")
```
