```python
#Aluno: Wagner Fedrigo Camargo
"""
NIVEL MEDIO

2. Elabore os códigos em Python dos problemas a seguir:
a. Maçãs verdes custam R$ 12,20 por quilo. Cada uma pesa em torno de
130g. Faça um programa que tenha como entrada a quantidade de
maçãs e mostre o valor em reais.
b. Ler as notas da 1a. e 2a. avaliações de um aluno. Calcular a
média aritmética simples e escrever a média calculada.
c. A jornada de trabalho semanal de um funcionário é de até 40
horas. Escreva um algoritmo que leia o número de horas
trabalhadas em cada semana, o salário por hora e escreva o
salário total do mês do funcionário (considere que o mês possua 4
semanas exatas).
"""
print ("==============================================")
print ("Letra A - Nível Médio - Wagner Fedrigo Camargo")
print ("==============================================")
a = int(input("Digite a quantidade de maças: "))
total = a * 0.13 * 12.20
total = "R$ {:,.2f}".format(total).replace(",", "X").replace(".", ",").replace("X", ".")
print('O valor das maças é', total,"\n")
print ("==============================================")
print ("Letra B - Nível Médio - Wagner Fedrigo Camargo")
print ("==============================================")
a = int(input("Digite a 1º nota: "))
b = int(input("Digite a 2º nota: "))
media = (a + b)/2
print("A média das notas é:", media,"\n")
print ("==============================================")
print ("Letra C - Nível Médio - Wagner Fedrigo Camargo")
print ("==============================================\n")
salHora = int(input("Digite o valor do Salário por Hora: "))
sem1 = int(input("Digite a quantidade de horas trabalhadas da 1º semana: (Máximo: 40 horas)"))
sem2 = int(input("Digite a quantidade de horas trabalhadas da 2º semana: (Máximo: 40 horas)"))
sem3 = int(input("Digite a quantidade de horas trabalhadas da 3º semana: (Máximo: 40 horas)"))
sem4 = int(input("Digite a quantidade de horas trabalhadas da 4º semana: (Máximo: 40 horas)"))
totalSal = (sem1 + sem2 + sem3 + sem4) * salHora
totalSal = "R$ {:,.2f}".format(totalSal).replace(",", "X").replace(".", ",").replace("X", ".")
print('O valor total do salário é', totalSal,"\n")
print ("==============================================")
```

```python
#Aluno: Wagner Fedrigo Camargo
"""
NIVEL AVANÇADO

3. Elabore os códigos em Python dos problemas a seguir:
a. A medida de eficiência de um carro é medida em km/l (quantos km o
carro anda com 1 litro de combustível). Faça um algoritmo que
calcule a eficiência de um carro tendo como entrada a distância
percorrida e a quantidade de combustível gasta em litros.

Exemplo:
distância percorrida = 300km
quantidade gasta = 20 litros
eficiência: 15km/l
"""
print ("=================================================")
print ("Letra A - Nível Avançado - Wagner Fedrigo Camargo")
print ("=================================================")
a = int(input("Digite a Distancia Percorrida: "))
b = float(input("Digite a Quantidade de Litros Gastos: "))
eficiencia = (a / b)
print("A eficiência do carro é:", round(eficiencia,2) ,"km/l\n")
"""
NIVEL AVANÇADO

b. Atualmente existem os carros flex, ou seja, podem ser abastecidos
com etanol ou gasolina. Geralmente um carro flex abastecido a
etanol tem uma eficiência menor do mesmo abastecido com gasolina.
Faça um algoritmo que tendo como entrada: eficiência do carro
abastecido à gasolina (em km/l), eficiência do carro abastecido
com etanol (em km/l) , preço da gasolina por litro e preço do
etanol por litro e mostre o valor do km rodado com cada um dos
combustíveis.

Exemplo:
eficiência com gasolina: 15km/l
preço da gasolina: R$ 4,50 por litro
valor do km rodado com gasolina: R$ 0,30

eficiência com etanol: 10km/l
preço do etanol: R$ 3,50 por litro
valor do km rodado com etanol: R$ 0,35
"""
print ("=================================================")
print ("Letra B - Nível Avançado - Wagner Fedrigo Camargo")
print ("=================================================")
efiGas = int(input("Digite a Eficiência com Gasolina: "))
precoGas = float(input("Digite o preço da Gasolina: "))
efiAlc = int(input("Digite a Eficiência com Etanol: "))
precoAlc = float(input("Digite o preço da Etanol: "))
kmGas = (precoGas / efiGas)
kmGas = "R$ {:,.2f}".format(kmGas).replace(",", "X").replace(".", ",").replace("X", ".")
kmAlc = (precoAlc / efiAlc)
kmAlc = "R$ {:,.2f}".format(kmAlc).replace(",", "X").replace(".", ",").replace("X", ".")
print("\nO valor do km rodado com gasolina é:", kmGas)
print("O valor do km rodado com etanol é:", kmAlc)
print ("=================================================\n")

"""
NIVEL AVANÇADO
c. Uma empresa deseja cadastrar em seu banco de dados os nomes de
candidatos à vagas de trabalho. Os nomes são recebidos dos
usuários e é necessário adotar um padrão. Foi decidido que os
nomes estariam todos em letras maiúsculas e sem espaços no início
e no final. Também é desejado que o primeiro nome, nome do meio e
sobrenome seja gravado de forma separada.

Exemplo:
Nome recebido: “José Antunes da Silva”
Dados a serem gravados:
Primeiro Nome: “José”
Nome do meio: “Antunes da”
Sobrenome: “Silva”
"""
print ("=================================================")
print ("Letra C - Nível Avançado - Wagner Fedrigo Camargo")
print ("=================================================")
nomeCompleto = str(input("DIgite o nome completo: ")).strip()
nomeCompleto = nomeCompleto.upper()
nomeCompleto = nomeCompleto.split()
print("Primeiro Nome:",nomeCompleto[0])
print("Nome do Meio:",nomeCompleto[1] , nomeCompleto[2])
print("Sobre Nome:",nomeCompleto[3])
print ("=================================================")

```
